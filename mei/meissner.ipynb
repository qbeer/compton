```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
down=np.loadtxt('hules.dat')
up=np.loadtxt('melegedes.dat')
```

```python
def utrans(x):
    return *x+
```

```python
plt.figure(figsize=(8, 6))
plt.plot(down[:,0], utrans(down[:,1]), "x", color="blue", label=u"hűlés")
plt.plot(up[:,0], utrans(up[:,1]), ".", color="red", label=u"melegedés", )
plt.xlim(82, 105)
plt.xlabel(u"hőmérséklet [K]", fontsize=14)
plt.ylabel(u"$-\Delta$ U [mV]", fontsize=14)
plt.legend(fontsize=14)
```

```python
def shift_p(x, s):
    return x+s
def shift_n(x, s):
    return x-s
```

```python
shift=0#0.325

down_s=shift_p(down[:,0], shift)
up_s= shift_n(up[:,0], shift)

plt.figure(figsize=(10, 8))
plt.plot(down[:, 0], utrans(down[:,1]), "x", color="blue", label=u"hűlés")
plt.plot(up[:, 0], utrans(up[:,1]), ".", color="red", label=u"melegedés", )
plt.axhline(y=utrans(0.04055), color='g', linestyle='-')
plt.axvline(x=92.52, color='g', linestyle='-')
plt.axvline(x=91.88, color='g', linestyle='-')
plt.xlim(89, 95)
plt.xlabel(u"hőmérséklet [K]", fontsize=14)
plt.ylabel(u"$-\Delta$ U [mV]", fontsize=14)
plt.legend(fontsize=14)
plt.grid(True)
```

```python

dTd=down[:-1, 0]-down[1:, 0]
dtd=down[1:, 2]-down[:-1, 2]
v_d=dTd/dtd

dTu=up[1:, 0]-up[:-1, 0]
dtu=up[1:, 2]-up[:-1, 2]
v_u=dTu/dtu
```
